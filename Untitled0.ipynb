```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.loc[penguins.bill_length_mm >= 42, 'bill_group'] = 'High'
penguins.loc[(penguins.bill_length_mm >35) & (penguins.bill_length_mm < 42), 'bill_group'] = 'Middle'
penguins.loc[penguins.bill_length_mm <= 35, 'bill_group'] = 'Low'
penguins.head()
```

```python
penguins.info()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue='species', size='island')
```

```python
sns.scatterplot(data=penguins, x="island", y="body_mass_g", hue='species', size='bill_length_mm')
```

```python
sns.PairGrid(penguins, hue='species').map(sns.scatterplot).add_legend()
```

```python
data = penguins.pivot_table(index='species',columns='island',values='body_mass_g')
sns.heatmap(data)
```
